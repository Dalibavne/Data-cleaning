```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset (space separated)
file_path = r"C:\Users\dalib\Documents\4) house Prediction Data Set.csv"
data = pd.read_csv(file_path, delim_whitespace=True, header=None)

# Step 2: Assign correct column names
data.columns = [
    'CRIM',     # per capita crime rate
    'ZN',       # proportion of residential land zoned
    'INDUS',    # proportion of non-retail business acres
    'CHAS',     # Charles River dummy variable
    'NOX',      # nitric oxides concentration
    'RM',       # average number of rooms per dwelling
    'AGE',      # proportion of owner-occupied units built before 1940
    'DIS',      # weighted distances to employment centres
    'RAD',      # index of accessibility to highways
    'TAX',      # property-tax rate
    'PTRATIO',  # pupil-teacher ratio
    'B',        # proportion of Black population
    'LSTAT',    # % lower status of population
    'MEDV'      # Median value of owner-occupied homes (Target)
]

print(data.head())   # verify data
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.index, data['MEDV'], color='blue', marker='o', linewidth=1)
plt.title("Median House Value across Records")
plt.xlabel("Record Index")
plt.ylabel("Median Value (MEDV)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
data.groupby('RAD')['MEDV'].mean().plot(kind='bar', color='orange')
plt.title("Average Median Value by RAD Index")
plt.xlabel("RAD (Accessibility Index)")
plt.ylabel("Average Median Value")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['RM'], data['MEDV'], alpha=0.6, color='green')
plt.title("Rooms vs Median Value")
plt.xlabel("Average Rooms per Dwelling (RM)")
plt.ylabel("Median Value (MEDV)")
plt.show()
```

```python
plt.figure(figsize=(6,6))
data['CHAS'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Houses near Charles River (CHAS)")
plt.ylabel("")
plt.show()
```
